## Предобработка данных

### Обработка пропусков и некорректных значений

```python
# подсчет пропущенных значений в датасете
data.isnull().sum()
```

```python
# процент пропущенных значений в датасете
data.isnull().sum() / data.shape[0] * 100
```

**Обработка пропусков в поле `floors_total`**

Переменная `floors_total` содержит информацию о количестве этажей в доме. Данная переменная будет использоваться для исследования, поэтому необходимо обработать пропуски в ней. Т.к. пропусков в этом поле немного (0.36%), можно предположить, что они появились по причине человеческого фактора, пользователи просто забыли заполнить поле, когда составляли объявление. 

```python
# удалим строки с пустым значением floors_total из датасета
data = data.dropna(subset=['floors_total'])
```

**Обработка нулевых значений в поле `rooms`**

```python
# посчитаем количество нулевых значений в поле rooms
data[data.rooms == 0]['rooms'].count()
```

```python
# нулевых значений в датасете немного, удалим строки со значением rooms = 0
data = data.query('rooms > 0')
```

```python
data[data.rooms == 0]['rooms'].count()
```

**Обработка пропусков в поле `living_area`**

```python
# Функция для заполнение пропусков в поле living_area средним значением для квартир с
# таким же количеством комнат (rooms)

def living_area_pass(rooms):
    mean_living_area = data.groupby('rooms')['living_area'].mean()
    return mean_living_area[rooms]
```

```python
# Заполнение пропусков в поле living_area
data.loc[data.living_area.isnull(), 'living_area'] = data['rooms'].apply(living_area_pass)
```

```python
# проверка количества пропусков в поле living_area
data[data.living_area.isnull()]['living_area'].count()
```

**Обработка пропусков в поле `kitchen_area`**

```python
# Функция для заполнения нулевых значений в поле kitchen_area средним значением для квартир с
# таким же количеством комнат (rooms)

def kitchen_area_pass(total_area_int):
    mean_kitchen_area = data.groupby('rooms')['kitchen_area'].mean()
    return round(mean_kitchen_area[total_area_int], 2)
```

```python
# Заполнение пропусков в поле kitchen_area
data.loc[data.kitchen_area.isnull(), 'kitchen_area'] = data['rooms'].apply(kitchen_area_pass)
```

```python
# проверка количества пропусков в поле kitchen_area
data[data.kitchen_area.isnull()]['living_area'].count()
```

**Обработка пропусков и аномальных значений в поле `ceiling_height`**

Посмотрим сколько аномальных значений в поле `ceiling_height` (высота потолков (м)), это значения меньше 2,3 м и больше 5 м.

```python
data[(data.ceiling_height < 2.3) | (data.ceiling_height > 5)]['ceiling_height'].count()
```

Пропуски и аномальные значения в поле `ceiling_height` (высота потолков) заполним наиболее часто встречающимся значением данной переменной, которое найдем с помощью метода **mode()**.

```python
data.loc[(data.ceiling_height.isnull()) | ((data.ceiling_height < 2.3) | (data.ceiling_height > 5)),
          'ceiling_height'] = float(data.ceiling_height.mode())
```

```python
# проверка количества пропусков в поле ceiling_height
data[data.ceiling_height.isnull()]['living_area'].count()
```

**Обработка пропусков в поле `is_apartment`**

Пропуски в поле `is_apartment` (апартаменты (булев тип)) заполним значением **False**, т.к. если поле не было заполнено, то скорее всего квартира не относится к апартаментам.

```python
data.loc[data.is_apartment.isnull(), 'is_apartment'] = False
```

**Обработка пропусков в поле `balcony`**

Пропуски в поле `balcony` (число балконов) заполним нулями, т.к. если поле не было заполнено, то скорее всего в квартире нет балкона.

```python
data.loc[data.balcony.isnull(), 'balcony'] = 0
```

**Обработка пропусков в поле `locality_name`**

Пропуски в поле `locality_name` (название населённого пункта) заполним значением **Unknown**.

```python
data.loc[data.locality_name.isnull(), 'locality_name'] = 'Unknown'
```

**Обработка пропусков в поле `days_exposition`**

Переменная `days_exposition` содержит информацию о количестве дней размещения объявления (от публикации до снятия). Данная переменная будет использоваться для исследования, поэтому необходимо обработать пропуски в ней. Можно предположить, что пропуски в данном поле появились, т.к. эти объявления были сформированы пользователем, но по какой-то причине так и не опубликованы. 

```python
# удалим строки с пустым значением days_exposition из датасета
data = data.dropna(subset=['days_exposition'])
```

```python
# подсчет пропущенных значений
data.isnull().sum()
```

- **чем ближе к центру города, тем выше цена за кв. метр**. Центр города - это более удобное и престижное место для проживания из-за близости объектов инфраструктуры, поэтому из-за высокого спроса цены здесь выше. 

- **чем больше комнат, тем ниже цена за кв. метр**. Стоит отметить, что такая зависимость справедлива для самых распространенных квартир - *одно-, двух- и трехкомнатных*. С увеличением числа комнат, цены за кв. метр могут быть самыми разными. Это происходит из-за низкого числа подобных предложений.

- **квартира на первом или последнем этаже обойдется дешевле**. Это связано с тем, что спрос на такие квартиры ниже, т.к. люди видят потенциальные проблемы в таком расположении жилья: на первом этаже из-за низко расположенных окон в квартире будет больше шума и пыли с улицы, из-за соседства с подвалом, могут завестись грызуны и тараканы; на последнем этаже проблемы могут возникнуть с протеканием крыши (особенно в старых домах), также могут быть проблемы с лифтом или вовсе его отсутствие.

- **у объявлений, опубликованных в выходные дни, цены ниже**. Объяснить эту особенность можно тем, что в рабочие дни объявления в основном публикуют агенства недвижимости, риелторы и застройщики, а в выходные дни - собственники жилья, которые нашли время вне работы, чтобы составить объявление. Цены на новое жилье от застройщиков выше чем на вторичку, цены от агенств недвижимости и риелторов выше, чем от собственников, т.к. в них заложена комиссия за работу. Т.о. объявления, опубликованные в субботу - это в основном вторичное жилье от собственников, поэтому цена за кв. метр ниже.

- **самый лучший месяц для покупки квартиры - июнь**. Как и любому другому рынку недвижимости характерна сезонность. Падение цен в июне объясняется тем, что приходит летнее затишье с отпусками и отдыхом. Запас предложений на рынке еще в избытке, но спрос падает, а за ним и цены на квартиры.

- **в 2015-2018 гг. квартиру можно было купить на 10% дешевле, чем в 2014 или в 2019 г**. Снижение цен на недвижимость в период 2015-2018 гг. можно объяснить снижением спроса, который обусловлен в свою очередь снижением уровня доходов населения и соответственно снижением платежеспособности.
