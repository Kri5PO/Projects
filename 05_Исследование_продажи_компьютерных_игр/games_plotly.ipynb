# Исследование продажи компьютерных игр

Интернет-магазин «Стримчик» продаёт по всему миру компьютерные игры. Из открытых источников доступны исторические данные о продажах игр, оценки пользователей и экспертов, жанры и платформы (например, Xbox или PlayStation). Необходимо выявить определяющие успешность игры закономерности. Это позволит сделать ставку на потенциально популярный продукт и спланировать рекламные кампании.

Представлены данные до 2016 года. Нужно отработать принцип работы с данными при планировании кампании на 2017-й. Неважно, прогнозируете ли вы продажи на 2017 год по данным 2016-го или же 2027-й — по данным 2026 года.

В наборе данных попадается аббревиатура ESRB (Entertainment Software Rating Board) — это ассоциация, определяющая возрастной рейтинг компьютерных игр. ESRB оценивает игровой контент и присваивает ему подходящую возрастную категорию, например, «Для взрослых», «Для детей младшего возраста» или «Для подростков».

### Изменение типа данных

Чтобы изменить тип поля `user_score` на **float64**, необходимо убрать строковые значения **tbd**. Заменим значение **tbd** на **None**

```python
games.loc[games.user_score == 'tbd', 'user_score'] = None
```

```python
# изменение типа поля user_score на float64
games.user_score = pd.to_numeric(games.user_score, errors='coerce')
```

На данном этапе нельзя изменить типы полей `Year_of_Release` (год выпуска), `Critic_Score` (оценки пользователей) на **int64**, т.к. в них присутствуют пропуски значений.

### Обработка пропусков и некорректных значений

```python
# замена значения K-A на E в поле rating
games.loc[games['rating'] == 'K-A', 'rating'] = 'E'
```

```python
# уникальные значения (исключая nan) в поле Rating
sorted(games.loc[games.rating.notnull(), 'rating'].unique())
```

```python
# удалим строки с пустым значением name, year_of_release, genre из датасета
games = games.dropna(subset=['name', 'year_of_release', 'genre'])
```

```python
# Изменение типа поля year_of_release на int
games['year_of_release'] = games['year_of_release'].astype('int64')
```

```python
# Заполнение пропусков в поле rating значением unknown
games.loc[games.rating.isnull(), 'rating'] = 'unknown'
```

```python
# проверка пропущенных значений в датасете
games.isnull().sum()
```

```python
games.name = games.name.str.lower()
```

```python
# соберем датасет с неявными дубликатами, у которых в наименовании игр присутствует слово sales
delete = games[games['name'].str.contains('sales') == True]
delete 
```

## Составление портрета пользователя каждого региона

Необходимо определить для пользователя каждого региона (NA, EU, JP):

- Самые популярные жанры (топ-5). Пояснить разницу.
- Влияет ли рейтинг ESRB на продажи в отдельном регионе?

Т.к. одни и те же расчеты необходимо делать для нескольких параметров исследования (платформы и жанры), рационально написать функции для формирования рабочего датасета и построения гистограмм. Это сократит количество строк кода в исследовании.

```python
# функция для формирования таблицы топ-5 позиций по регионам
def top_by_regions(data, parameter):
    na = data.sort_values("na_sales", ascending=False).drop(["eu_sales", "jp_sales"], axis=1).head() # Сев. Америка
    eu = data.sort_values("eu_sales", ascending=False).drop(["na_sales", "jp_sales"], axis=1).head() # Европа
    jp = data.sort_values("jp_sales", ascending=False).drop(["eu_sales", "na_sales"], axis=1).head() # Япония

    datasets = [na, eu, jp]
    regions = ['Сев.Америка', 'Европа', 'Япония']
    for i in range(len(datasets)):
        datasets[i].columns = [parameter, 'sales'] # переименование колонок
        datasets[i]['region'] = regions[i]         # добавление колонки с названием региона
    work_data = pd.concat([na, eu, jp], ignore_index=True) # соединение таблиц
    
    return work_data
```

```python
# функция для построения гистограммы для топ-5 позиций по регионам
def top_by_regions_hist(data, parameter, title, xaxis_title):
    data = top_by_regions(data, parameter)
    fig = px.histogram(data, x=parameter, y="sales",
             color='region', barmode='group',
             height=450, width=900, text_auto='.3s')
    fig.update_layout(title=title, title_font_size=16,
                  xaxis_title=xaxis_title, xaxis_title_font_size=12,
                  yaxis_title="продажи, млн. копий", yaxis_title_font_size=12,
                  legend_title_text='Регионы продаж:')
    fig.update_traces(textfont_size=12, textangle=0, textposition="outside", cliponaxis=False)
    fig.show("svg")
```

## Проверка гипотез

Для определения типа распределения выборок необходимо построить гистограммы и графики QQ-plot - для оценки степени отклонения данных от нормального распределения.

Чтобы проверить гипотезу о равенстве среднего двух генеральных совокупностей по взятым из них выборкам, будем использовать **t-тест Студента** - метод ***stats.ttest_ind()***. Метод содержит необязательный параметр **equal_var**, задающий считать ли равными дисперсии выборок:
- **equal_var = True** - считать дисперсии выборок равными (по умолчанию),
- **equal_var = False** - считать дисперсии выборок неравными.

Минимальный уровень значимости **p-value**, на котором нулевая гипотеза может быть отвергнута, примем равным **0,05**, это указывает на вероятность справедливости гипотезы в 5%

```python
# функция для построения гистограмм распределения выборок
def histogram(data, title):
    fig = px.histogram(data)
    fig.update_layout(title=title, title_font_size=16,
                      xaxis_title="оценки пользователей", xaxis_title_font_size=12,
                      yaxis_title="частота", yaxis_title_font_size=12,
                      autosize=False, width=600, height=400, showlegend=False)
    fig.show("svg")
```

```python
# функция для проверки гипотез
def test(data1, data2, equal_var):
    alpha = 0.05 # критический уровень статистической значимости
    results = stats.ttest_ind(data1, data2, equal_var=equal_var)
    print('p-значение:', results.pvalue)
    
    if results.pvalue < alpha:
        print("Отвергаем нулевую гипотезу")
    else:
        print("Не получилось отвергнуть нулевую гипотезу")
```

### Средние пользовательские рейтинги платформ Xbox One и PC одинаковые

Нулевая гипотеза H₀: **Средние пользовательские рейтинги платформ Xbox One и PC одинаковые**.

Альтернативная гипотеза H₁: **Средние пользовательские рейтинги платформ Xbox One и PC разные**.

```python
data_xone = games.loc[(games.platform == 'XOne') & (games.user_score.notnull()), 'user_score']
data_pc = games.loc[(games.platform == 'PC') & (games.user_score.notnull()), 'user_score']
```

```python
histogram(data=data_xone, title="Распределение оценок пользователей платформы Xbox One")
histogram(data=data_pc, title="Распределение оценок пользователей платформы PC")
```

```python
# график QQ-plot для выборки data_xone
stats.probplot(data_xone, dist="norm", plot=pylab)
pylab.show()
# график QQ-plot для выборки data_pc
stats.probplot(data_pc, dist="norm", plot=pylab)
pylab.show()
```

```python
print('Размер выборок:')
print('data_xone -', len(data_xone))
print('data_pc -', len(data_pc))
```

По виду графиков QQ-plot и виду гистограмм можно сделать вывод, что распределение выборок **data_xone** и **data_pc** не является нормальным. Однако, в такой ситуации t-тест Студента применим для проверки гипотез при большом размере выборок (>30). В нашем случае размер выборок превышает значение 30.

Далее посмотрим на дисперсии по выборкам для определения значения параметра **equal_var** метода **stats.ttest_ind()**

```python
print('Дисперсия выборки по Xbox One:', round(np.var(data_xone), 1))
print('Дисперсия выборки по PC:', round(np.var(data_pc), 1))
print("Отличие дисперсий: {:.1%}".format(np.var(data_pc)/np.var(data_xone) - 1))
```

Дисперсии значительно отличаются, применим параметр **equal_var=False**

```python
test(data1=data_xone, data2=data_pc, equal_var=False)
```

В результате проверки нулевая гипотеза была отвергнута, p-значение очень мало, значит можно говорить о справедливости альтернативной гипотезы, а именно - **Средние пользовательские рейтинги платформ Xbox One и PC разные**
